```python
# Python program to create
# a pdf file


from fpdf import FPDF
#initial thresholds for the games
math_t = 20
block_t = 20
tictaktoe_t = 4
memory_t = 4

#input scores
math = 19
block = 21
tictaktoe = 3
memory = 5
# save FPDF() class into a
# variable pdf
pdf = FPDF()

# Add a page
pdf.add_page()

# set style and size of font
# that you want in the pdf
pdf.set_font("Arial", size = 14)

name = "Samyuktha"
# create a cell
pdf.cell(200, 10, txt = "Developmental report for "+name,
		ln = 1, align = 'C')

# add another cell
pdf.set_font("Arial", size = 12)
pdf.cell(200, 10, txt = "This report is generated based on the responses the parent gave to the form shared.",
		ln = 2, align = 'C')

pdf.set_font("Arial", size = 11)
# Math game
if math>math_t:
  
  pdf.cell(200, 10, txt = "Math Game: The score achieved by the kid is "+str(math)+". This is a form of improvement in the mathematical ability of the child.",
		ln = 3, align = 'J')
else:
    pdf.cell(200, 10, txt = "Math Game: The score achieved by the kid is "+str(math)+". The mathematical skill of the child can be improved by practice.",
		ln = 3, align = 'J')

# Block game
if block>block_t:
    pdf.cell(200, 10, txt = "Block Game: The score achieved by the kid is "+str(block)+". This is a form of improvement in the reflexiveness of the child.",
		ln = 4, align = 'J')
    block=block_t
else:
    pdf.cell(200, 10, txt = "Block Game: The score achieved by the kid is "+str(block)+". The reflexiveness of the child can be improved by practice.",
		ln = 4, align = 'J')

# Memory game
if memory>memory_t:
    pdf.cell(200, 10, txt = "Memory Game: The score achieved by the kid is "+str(memory)+". This is a form of improvement in the memory skill of the child.",
		ln = 5, align = 'J')
    memory=memory_t
else:
    pdf.cell(200, 10, txt = "Memory Game: The score achieved by the kid is "+str(memory)+". The memory skill of the child can be improved by practice.",
		ln = 5, align = 'J')

# Tic-tak-toe game
if tictaktoe>tictaktoe_t:
    pdf.cell(200, 10, txt = "Tic-Tak-Toe: The score achieved by the kid is "+str(tictaktoe)+". This is a form of improvement in the strategical skill of the child.",
		ln = 6, align = 'J')
    tictaktoe_t = tictaktoe
else:
    pdf.cell(200, 10, txt = "Tic-Tak-Toe: The score achieved by the kid is "+str(tictaktoe)+". The strategical skill of the child can be improved by practice.",
		ln = 6, align = 'J')

# save the pdf with name .pdf
pdf.output(name+".pdf")
```
